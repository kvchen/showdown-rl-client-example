# Pokemon Showdown RL Client Example

```python
import requests
import pprint

pp = pprint.PrettyPrinter(indent=4)
```

```python
HOST = "http://localhost"
PORT = 3000

BASE_URL = f"{HOST}:{PORT}/game"
```

```python
def create_battle():
    url = f"{BASE_URL}/new"
    return requests.post(url).json()

def do_move(battle_id, move_type, message):
    url = f"{BASE_URL}/{battle_id}/move"
    return requests.post(url, data={"moveType": move_type, "message": message}).json()

def cleanup(battle_id):
    url = f"{BASE_URL}/{battle_id}"
    return requests.delete(url)
```

```python
def print_log(log):
    print('\n'.join(log))
```

```python
game = create_battle()
battle_id, battle = game["battleID"], game["battle"]

print(battle_id)
# pp.pprint(battle)
```

```python
ended, counter = False, 0

while not ended:
    next_battle = do_move(battle_id, "p1" if counter % 2 == 0 else "p2", "default")
    battle_id = next_battle["battleID"]
    ended = next_battle["battle"]["ended"]
    counter += 1

pp.pprint(next_battle["battle"])
```
